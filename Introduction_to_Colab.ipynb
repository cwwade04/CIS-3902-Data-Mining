```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv("data.csv")
# To display the top 5 rows
df.head(5)
```

```python
df.tail(5)                        # To display the botton 5 rows
```

```python
df.dtypes
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()      # Used to count the number of rows
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

```python
Q1 = df.select_dtypes(include=np.number).quantile(0.25)
Q3 = df.select_dtypes(include=np.number).quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
Q1 = df.select_dtypes(include=np.number).quantile(0.25)
Q3 = df.select_dtypes(include=np.number).quantile(0.75)
IQR = Q3 - Q1

# Filter df to only include numeric columns for comparison
numeric_df = df.select_dtypes(include=np.number)

# Apply outlier removal logic to the numeric DataFrame
df_filtered = numeric_df[~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Reassign the filtered numeric data back to the original DataFrame
df[numeric_df.columns] = df_filtered

df.shape
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
plt.figure(figsize=(10,5))
numeric_df = df.select_dtypes(include=np.number) # Select only numeric columns for correlation
c= numeric_df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
